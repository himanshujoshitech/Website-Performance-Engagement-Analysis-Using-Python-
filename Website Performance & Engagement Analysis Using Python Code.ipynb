```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# | Library    | Alias | Purpose                     |
# | ---------- | ----- | --------------------------- |
# | pandas     | pd    | Data handling (tables)      |
# | numpy      | np    | Numerical calculations      |
# | matplotlib | plt   | Basic graphs                |
# | seaborn    | sns   | Advanced & beautiful graphs |

# Importing Matplotlib Library
# Matplotlib is a Python library used for data visualization.
# The pyplot module helps in creating various charts and graphs easily.
# The alias plt is used to simplify plotting commands.

# Importing Seaborn Library
# Seaborn is a Python data visualization library based on Matplotlib.
# It is used to create attractive and informative statistical graphics.
# The alias sns is commonly used for convenience.
```

```python
df = pd.read_csv("data.csv")

# The pd.read_csv() function is used to load a CSV file into a pandas DataFrame.
# A DataFrame is a two-dimensional data structure consisting of rows and columns.
# The dataset is stored in the variable df for further analysis.
```

```python
df.head()

# The head() function is used to display the first few rows of a DataFrame.
# By default, it displays the first 5 rows of the dataset.
# This helps in quickly understanding the structure and content of the data.
```

```python
df.columns = df.iloc[0]
df.head()

# ðŸ”¹ iloc

# Stands for integer location

# Used to select data using row numbers

# Counting starts from 0

# ðŸ”¹ [0]

# Means first row

# Python indexing always starts from 0

# The df.columns attribute is used to define column headers of a DataFrame.
# df.iloc[0] selects the first row of the dataset using integer-based indexing.
# This line assigns the first row as column names when headers are present inside the data.
```

```python
df = df.drop(index = 0).reset_index(drop = True)

# The drop() function is used to remove unwanted rows from the DataFrame.
# Here, the first row is removed because it was used as column headers earlier.
# The reset_index() function is used to reset row numbers, and drop=True prevents the old index from being added as a column.



df.columns = ["channel group", "DateHour", "Users", "Sessions", "Engaged sessions", "Average engagement time per session", "Engaged sessions per user", "Events per session", "Engagement rate", "Event count"]

# Renaming Columns Manually
# The df.columns attribute is used to assign custom column names to the DataFrame.
# A list of column names is provided in the correct order.
# This step improves data readability and makes analysis easier.
```

```python
df.head()
```

```python
df.info()

# Dataset Information
# The info() function provides a concise summary of the DataFrame.
# It displays the number of rows and columns, column names, non-null values, and data types.
# This function is useful for identifying missing values and understanding the structure of the dataset.
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format = "%Y%m%d%H", errors = 'coerce')

# Converting Date Column to Datetime Format
# The pd.to_datetime() function is used to convert a column into datetime format.
# The format parameter specifies the structure of the date values.
# The errors='coerce' argument ensures that invalid values are converted to NaT instead of raising an error.
# Converting dates to datetime format enables time-based analysis.


df.head()

```

```python
numeric_cols = df.columns.drop(["channel group", "DateHour"])

# Selecting Numeric Columns
# The df.columns attribute returns all column names in the DataFrame.
# The drop() function is used to exclude non-numeric columns such as text and datetime fields.
# The remaining column names are stored for numeric conversion.

    
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors = 'coerce')

# Converting Columns to Numeric Data Type
# The apply() function is used to apply a conversion function across multiple columns.
# pd.to_numeric() converts values into numeric format.
# The errors='coerce' parameter ensures invalid values are converted to NaN.



df["Hour"] = df["DateHour"].dt.hour

# Extracting Hour from Datetime Column
# The .dt accessor is used to work with datetime values in pandas.
# The .hour attribute extracts the hour component from each datetime value.
# The extracted values are stored in a new column named Hour.
```

# Engaged VS Non-Engaged Sessions

```python
df.head()
```

```python
session_df=df.groupby("channel group")[["Sessions", "Engaged sessions"]].sum().reset_index()

# The groupby() function is used to group data by channel group.
# The sum() function calculates total sessions and engaged sessions for each channel.
# The reset_index() function restores the grouped column as a regular column.

# What happens WITHOUT reset_index()

# When you use groupby() and sum(), the result looks like this:

#                       Sessions  Engaged sessions
# channel group                                    
# Email                     125              105
# Social Media              250              200
# Direct                    180              150
# Notice:

# "channel group" is indented/pushed to the left
# It's not a normal column - it's called an "index"
# An index is like a special label row, not regular data



# What happens WITH reset_index()
# After you add reset_index(), it becomes:

#   channel group  Sessions  Engaged sessions
# 0         Email       125              105
# 1  Social Media       250              200
# 2        Direct       180              150


# Notice:

# "channel group" is now a regular column (same level as Sessions and Engaged sessions)
# There's a new simple number index (0, 1, 2) on the left
# Everything is aligned nicely





session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]

# Calculating Non-Engaged Sessions
# A new column named Non-Engaged is created by subtracting engaged sessions from total sessions.
# This represents sessions where users did not meaningfully engage.




session_df_melted = session_df.melt(
    id_vars="channel group",
    value_vars=["Engaged sessions", "Non-Engaged"]
)

# Reshaping Data Using Melt
# The melt() function is used to convert data from wide format to long format.
# The id_vars parameter specifies identifier columns, while value_vars defines columns to be transformed.
# This format is suitable for comparative visualizations.

# Why do we use melt?
# Main reason: To make charts/graphs easier!
# Many charting libraries (like Plotly, Seaborn, Matplotlib) work BETTER with long format data.
# Example - Making a grouped bar chart:
# If you want a chart showing Engaged vs Non-Engaged sessions for each channel:

# Wide format (before melt): Hard to create grouped bars
# Long format (after melt): Easy! Just say "x=channel group, y=value, color=variable"


# Real-world analogy:
# Imagine a spreadsheet tracking test scores:
    
# Wide format (before melt):
# Student    Math  Science  English
# Alice      85    90       88
# Bob        78    82       85

# Long format (after melt):
# Student   Subject   Score
# Alice     Math      85
# Alice     Science   90
# Alice     English   88
# Bob       Math      78
# Bob       Science   82
# Bob       English   85

# Which is easier to:

# Make a chart comparing subjects? â†’ Long format!
# Filter by subject? â†’ Long format!
# See overall patterns? â†’ Long format!




# WHY THIS STEP IS VERY IMPORTANT (BIG PICTURE)

# You just prepared data for:
# âœ” Stacked bar charts
# âœ” Engagement comparison
# âœ” Channel performance analysis

# This is real-world analytics transformation 
```

```python
plt.figure(figsize=(8,5))

# Creating the Plot Area
# The plt.figure() function creates a new figure with a specified size to improve readability.


sns.barplot(data=session_df_melted, x="channel group", y="value", hue="variable")

# Bar Chart for Engaged vs Non-Engaged Sessions
# The sns.barplot() function is used to compare engaged and non-engaged sessions across channel groups.
# The hue parameter separates session types using different colors for easy comparison.



plt.title("Engaged VS Non-Engaged Sessions")

# Adding Plot Title
# The plt.title() function adds a descriptive title to the chart.

    

plt.xticks(rotation=45)

# Rotating X-Axis Labels
# The plt.xticks() function rotates x-axis labels to improve readability.


plt.show()

# Displaying the Plot
# The plt.show() function renders and displays the visualization.
```

# Traffic by Hour and Channel

```python
heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack().fillna(0) 

# Preparing Data for Heatmap
# The data is grouped by hour and channel group to calculate total sessions.
# The unstack() function reshapes the grouped data into a matrix format suitable for heatmaps.
# Missing values are replaced with zero using fillna(0).

# unstack

# Method/function (THIS IS THE KEY PART!)
# Converts one of the grouped columns into column headers
# Transforms from long format to wide format
# Opposite of melt!

# Visual Example:
# BEFORE unstack (after groupby and sum):
# Hour  channel group    Sessions
# 0     Email                 50
# 0     Social Media         100
# 0     Direct                75
# 1     Email                 60
# 1     Social Media         120
# 1     Direct                80
# ...

# AFTER unstack:
# channel group   Email  Social Media  Direct
# Hour                                        
# 0                  50           100      75
# 1                  60           120      80
# 2                  45           110      70
# ...



plt.figure(figsize=(12,6))

# Creating Figure for Heatmap
# A larger figure size is used to clearly display the heatmap.


sns.heatmap(
    heatmap_data,
    cmap="YlGnBu",
    linewidths=.5,
    annot=True,
    fmt='.0f'
)

# Creating Heatmap Visualization
# The heatmap visualizes session intensity using color gradients.
# Annotations are enabled to display exact values, and color mapping enhances pattern recognition.

# sns

# This is a library/module name
# Short for Seaborn (a Python visualization library)
# You imported it earlier with: import seaborn as sns


# . (dot)

# Dot operator
# Means "use a function from the sns library"


# heatmap

# This is a function/method from Seaborn
# Creates a heatmap visualization
# A heatmap shows data as colors in a grid


# ( (opening parenthesis)

# Starts the arguments/parameters for the heatmap function
# Everything inside tells the function HOW to draw the heatmap


# heatmap_data

# This is the variable you created earlier
# Contains your grid data (Hours Ã— Channel Groups with session counts)
# This is the MAIN data that will be visualized


# , (comma)

# Separator
# Separates different parameters


# cmap

# Parameter name
# Short for "color map"
# Controls what colors the heatmap uses


# = (equals)

# Assignment for parameter
# Sets the value for cmap


# "YlGnBu"

# A string (text in quotes)
# Name of a specific color scheme
# "YlGnBu" = Yellow-Green-Blue
# Low values = Light Yellow
# Medium values = Green
# High values = Dark Blue

# Other popular options:

# "coolwarm" = Blue to Red
# "Reds" = White to Dark Red
# "Greens" = White to Dark Green


# , (comma)

# Separator


# linewidths

# Parameter name
# Controls the width of lines between cells in the heatmap
# Creates grid lines separating each box


# = (equals)

# Sets the value for linewidths


# .5

# A number (decimal)
# Same as 0.5
# The thickness of the grid lines
# Higher number = thicker lines
# 0 = no lines between cells


# , (comma)

# Separator


# annot

# Parameter name
# Short for "annotations"
# Controls whether to show numbers inside each cell


# = (equals)

# Sets the value for annot


# True

# A boolean value (True or False)
# True = YES, show the numbers inside each cell
# False = Don't show numbers, just colors


# , (comma)

# Separator


# fmt

# Parameter name
# Short for "format"
# Controls HOW the numbers are displayed


# = (equals)

# Sets the value for fmt


# '.0f'

# A string (format specification)
# Let me break this down:

# . (dot in the string)

# Indicates a decimal number format

# 0

# Number of decimal places to show
# 0 = no decimal places (whole numbers only)
# 1 = one decimal place (e.g., 123.4)
# 2 = two decimal places (e.g., 123.45)

# f

# Means "float" (decimal number)
# Fixed-point notation

# Examples:

# '.0f' â†’ 123 (no decimals)
# '.1f' â†’ 123.4 (one decimal)
# '.2f' â†’ 123.45 (two decimals)


# ) (closing parenthesis)

# Ends all the parameters for the heatmap function



plt.title("Traffic by Hour and Channel")





plt.xlabel("Channel Group")





plt.ylabel("Hour of Day")




plt.show()

```
